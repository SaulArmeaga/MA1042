```python
#Importamos librerias necesarioas

#Libreria nescesaria para poder leer la informacion como un dataset
import pandas as pd 
#Libreria para graficar
import matplotlib.pyplot as plt
```

```python
#Carga el dataset
df_limpieza = pd.read_excel('/content/Actuaciones_limpieza_urbana_2022.xlsx')
df_limpieza
```

```python
#Hace el conteo de cuantos NaN tiene el dataset
df_limpieza.isna().sum()
```

```python
#Seleccion de la informacion con la que trabajaremos
df_limpieza_info = df_limpieza.iloc[:][['MES','Kg Recogida de muebles','Kg Recogida Residuos Viarios','RESTO            ']]
df_limpieza_info
```

```python
# De la informacion que seleccionamos, almacenamos aquella que deseamos 
# Analizar 
toda_info= df_limpieza_info.iloc[:,1:]

#M
for i in toda_info:
  print("******************************************")
  print(f'\t{i}')
  print("******************************************")
  print(f'Su media es              : {toda_info[i].median():15.3f}')
  print(f'Su valor maximo es       : {toda_info[i].max():15.3f}')
  print(f'Su valor minimo es       : {toda_info[i].min():15.3f}')
  print(f'Su rango es              : {toda_info[i].max()-toda_info[i].min():15.3f}')
  print(f'La varinza es            : {round(toda_info[i].var(),3):15.3f} ')
  print(f'La desviación estándar es: {round(toda_info[i].std(),3):15.3f}')
  print(f'')
```

```python
nombres = ["Kg Recogida de muebles","Kg Recogida Residuos Viarios","RESTO"]
colores = ['orange','red','green']
cont = 0
for i in toda_info:
  #histobrama
  plt.figure()
  plt.hist(toda_info[i],bins = 18, color=colores[cont])
  plt.title(f'Histograma de {nombres [cont]}')
  plt.xlabel(f'{nombres[cont]}')
  plt.ylabel('Frecuencia')
  plt.show()

  #Caja bigote
  plt.figure()
  plt.title(f'Caja bigote de {nombres [cont]}')
  plt.boxplot(toda_info[i])
  plt.show()
  cont+=1
```
